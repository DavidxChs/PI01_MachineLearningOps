#### -  Desanidar las columnas que lo necesiten<br>
Creo una función para convertir los valores de cada fila en un diccionario o en una lista.<br>
Utilizo el método `ast.literal_eval()` que convierte los valores en diccionario o lista
- sí aparencen los valores entre [] se convierte en lista
- si están entre {} se convierte en diccionario.

```python
def to_dict_list(fila):
    if pd.isna(fila): #si la fila es NaN no se devuelve ningún valor
        return np.nan
    else:
        return ast.literal_eval(fila) # Convierte en un diccionario o en lista
```

Aplico la función **to_dict** para convertir en un diccionario los elementos de la coulumna **belongs_to_collection**

```python
df['belongs_to_collection'] = df['belongs_to_collection'].apply(to_dict_list)
```

Convirto la columna **belongs_to_collection** en un DataFrame

Para la desanidar las demás columnas que están en una lista de diccionarios creo una función para obtener valores que están en una lista de diccionarios

```python
def obtener_valores(lista):
    NewList = [] # Lista en la que se colocarán los valores extraidos
    NewStr = '' # Candena vacía en la que se almacenarán los elementos de NewList
    if (type(lista) == list):
        for diccionario in lista:
            for clave, valor in diccionario.items():
                # Cuando la clave del diccionario sea 'name' se agrega es valor a la lista "NewList"
                if (clave == 'name'):
                    NewList.append(valor)
    else:
        NewList.append("")
    for item in NewList:
        if NewList.index(item) == 0: # Si es el primer elemento de la lista solo colocar el valor del item
            NewStr = item
        else:
            NewStr = NewStr + ', ' + item # Si no es el primer elemento colocar una coma antes
    return NewStr
```

Convieto las columnas en listas de diccionarios y después reemplazo los datos desanidados en el DataFrame principal.

```python
# Convierto las columnas en listas de diccionarios.
df['genres'] = df['genres'].apply(to_dict_list) # Se convierten las columnas en listas de diccionarios
df['production_companies'] = df['production_companies'].apply(to_dict_list)
df['production_countries'] = df['production_countries'].apply(to_dict_list)
df['spoken_languages'] = df['spoken_languages'].apply(to_dict_list)
```

```python
# Reemplazo los valores de los datos desanidados en el DataFrame principal
df['genres'] = df['genres'].apply(obtener_valores)
df['production_companies'] = df['production_companies'].apply(obtener_valores)
df['production_countries'] = df['production_countries'].apply(obtener_valores)
df['spoken_languages'] = df['spoken_languages'].apply(obtener_valores)
df.head(1)
```

#### - Cambiar formato de fechay y agregar la columna **release_year**<br>

Hago una función para cambiar los valores tipo `string` a un valor tipo `datetime` con un formato específico.

```python
# Función para cambiar los string a tipo datetime
def todate(strdates):
    formato = "%Y-%m-%d" # Formato único para todas las fechas
    return datetime.strptime(strdates, formato).date() # Se convierte en datetime
```

Veré cuántas filas cumplen con el formato de fecha, y visualizaré las filas que no cumplan con el formato de fecha

```python
formato_date = r'\d{4}-\d{2}-\d{2}'
# Aquí se marcan las fechas que sí cumplen con el formato
match_date = df['release_date'].str.match(formato_date)
print(df[match_date].shape[0], 'filas sí cumplen con el formato')
print(df.shape[0] - df[match_date].shape[0], 'filas no cumplen con el formato')
```

```python
# Reviso las filas que no cumplen con el formato
no_match = ~match_date
no_matchdf = df[no_match]
no_matchdf
```

#### - Crear la columna con el retorno de inversión, llamada **return**<br>
La columna **return** es creada con los campos **revenue** y **budget**, dividiendo estas dos últimas ``revenue / budget``, si no hay datos disponibles para calcularlo, tomará el valor 0

```python
# Antes de realizar la operación para crear la nueva columna,
# verifico el tipo de datos de cada campo del DataFrame
df.info()
```

```python
df['budget'] = df['budget'].astype(float)
df['return'] = df['revenue'] / df['budget']
df['return'] = pd.Series([round(value, 2) for value in df['return']])
```

```python
# Reviso como quedó la columna "return" poniendo de referencia los campos "revenue" y "budget"
df[['revenue', 'budget', 'return']]
```

- Los valores NaN en la tabla return se deben a que no hay información ni en el campo **revenue**, ni en **budget**, por lo tanto sustituiré el NaN por cero
- Los valores infinitos son debidos a la división del campo **revenue** entre un valor **budget** igual a cero, es decir, sin información, por lo anterior asignaré también un valor de cero en el campo **return** en el caso correspondiente

```python
# Reemplazo los valores NaN por cero, y los infinitos por NaN
df['return'] = df['return'].replace([np.nan, np.inf], [0, 0])
```

```python
#Función para obtener el nombre de los directores a partir de una lista de diccionarios
def obtain_values(lista):
    NewList = [] #Lista en la que se colocarán los valores extraidos
    NewStr = '' # Candena vacía en la que se almacenarán los elementos de NewList
    if (type(lista) == list):
        for diccionario in lista:
            director = False #En cada diccionario indica que no se ha encontrado al director
            for clave, valor in diccionario.items():
                if (clave == 'job' and valor == 'Director'):
                    director = True #Si en un diccionario se encuentra el director, director = True
                if (clave == 'name' and director == True):
                    NewList.append(valor)
    else:
        NewList.append("")
    for item in NewList:
        if NewList.index(item) == 0: # Si es el primer elemento de la lista solo colocar el valor del item
            NewStr = item
        else:
            NewStr = NewStr + ', ' + item # Si no es el primer elemento colocar una coma antes
    return NewStr
```
